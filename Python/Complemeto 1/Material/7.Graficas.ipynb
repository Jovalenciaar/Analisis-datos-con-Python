## Graficando varios precios en la misma gráfica

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,5))

# Empiezo dibujando solo Open vs fecha
ax = NFLX.Open.plot(legend=True)
ax = NFLX.Close.plot(legend=True)

# Pongo las etiquetas a la gráfica

ax.set_xlabel('Fecha')
ax.set_ylabel('Prices')
```

## graficando con diferente escala

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,5))

# Empiezo dibujando solo Open frente fecha
ax = NFLX.Open.plot(legend=True)
ax = NFLX.Close.plot(legend=True)

# Añado Volume, indicando se usa un eje y secundario
NFLX.Volume.plot(secondary_y=True, style='g', legend=True)

# Pongo las etiquetas a la gráfica
ax.set_xlabel('Fecha')
ax.set_ylabel('Prices')
ax.right_ax.set_ylabel('Volume')
```

## Graficando multiples stocks

```python
import pandas_datareader as pdr
import datetime
```

```python
start = datetime.datetime(2000, 10, 2)
end = datetime.datetime(2021, 10, 2)
```

```python
stocks = ["AMZN","AAPL",'MSFT', 'INTC',"IBM", "CSCO"]
stocks_df = pdr.DataReader(stocks, 'yahoo', start, end)

stocks_df = stocks_df['Adj Close']
stocks_df.head()
```

```python
stocks = ["AMZN","AAPL",'MSFT', 'INTC',"IBM", "CSCO"]

fig = plt.figure(figsize=(5, 12))
for i, sc in enumerate(stocks):
    fig.add_subplot(3, 2, i+1)
    ax = stocks_df[sc].plot(kind='hist')
    ax.set_ylabel('')
    ax.set_title(sc)
    ax.grid()
   
```

```python
col=["IBM","CSCO"]
fig, ax = plt.subplots(figsize=(12,5))
stocks_df[col].plot.kde(ax=ax, legend=False, title='Histogram: IBM vs. CSCo')
stocks_df[col].plot.hist(density=True, ax=ax)
ax.set_ylabel('Probability')
ax.grid(axis='y')
ax.set_facecolor('#d8dcd6')
```
