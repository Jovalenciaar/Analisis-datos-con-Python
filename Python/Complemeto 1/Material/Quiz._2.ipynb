```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
start = datetime.datetime(2000,10,2)
end = datetime.datetime(2023,10,2)
df = yf.download(["KO","PEP","K","NSRGY","CPB","SYY"], start=start, end=end)

```

```python
stock_df = df["Adj Close"]
```

#### Comparar la medi con la mediana

```python
stock_df.mean()==stock_df.median()
```

#### Cual es el precio de los tocks el 4 de octubre.

```python
stock_df.loc["2010-10-04"].round(2)
```

#### Histogramas

```python
stock = ["KO","PEP","K"]

fig = plt.figure(figsize=(10, 10))
for i, sc in enumerate(stock):
    fig.add_subplot(2, 3, i+1)
    ax = stock_df[sc].plot(kind='hist')
    ax.set_ylabel("")
    ax.set_title(sc)
    ax.grid()
```

```python
stock = ["CPB","SYY"]

fig = plt.figure(figsize=(10, 10))
for i, sc in enumerate(stock):
    fig.add_subplot(2, 3, i+1)
    ax = stock_df[sc].plot(kind='hist')
    ax.set_ylabel("")
    ax.set_title(sc)
    ax.grid()
```
