```python
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
```

```python
start = datetime.datetime(2000,10,2)
end = datetime.datetime(2023,10,2)
df = yf.download(['KO', 'PEP','K','NSRGY','CPB','SYY'], start=start, end=end)
```

```python
df_adj = df['Adj Close']
df_adj
```

```python
df_adj['NSRGY'].plot(kind='hist', bins=10)
plt.show()
```

```python
stock = ['KO', 'PEP','K','NSRGY','CPB','SYY']

fig = plt.figure(figsize=(10,10))
for i, sc in enumerate(stock):
    fig.add_subplot(3,2, i+1)
    ax = df_adj[sc].plot(kind='hist')
    ax.set_ylabel("")
    ax.set_title(sc)
    ax.grid()
```
