 ## Precios historicos

```python
Amazon.history(period="max")
```

### Definir nuestras propias fechas

```python
start = dt.datetime(2012,5,31)
end = dt.datetime(2013,1,30)

Amazon.history(start=start, end=end)
```

### Observación

```python
data = Amazon.history(interval="1d", start="2022-08-03", end="2023-02-10")
data.head(4)
```

### Descargar multiples tickers

### Descargar multiples tickers con inicio y final

```python
data_3 =yf.download(["GOOGL", "META"], start="2021-12-10", end="2021-12-30", group_by="ticker")
data_3.head(4)
```

```python
data_3["GOOGL"]
```
