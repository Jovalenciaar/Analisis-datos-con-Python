**4.b Para las siguientes entidades bancarias que cotizan en el IBEX 35 Calcular a nivel anual, en la ventana de tiempo del "2017-10-01" al "2024-10-02"  El ratio de Sharpe y  EL ratio de Sortino**

Sugerencia: revise como se calcula el libre de riesgo

* Banco Sabadell - SAB.MC
* Banco Santander - SAN.MC
* Bankinter - BKT.MC
* BBVA - BBVA.MC
* CaixaBank - CABK.MC
* Unicaja Banco - UNI.MC

```python
lista  = ['SAB.MC', 'SAN.MC','BKT.MC','BBVA.MC','CABK.MC','UNI.MC']
```

```python
stock = yf.download(lista, start = "2017-10-01", end = "2024-10-02")
stock_adj = stock[["Adj Close"]]
stock_adj
```

```python
stock_anual = stock_adj.resample('Y').apply(lambda x: (x.pct_change() + 1).product()-1)
stock_anual
```

```python
ret_anual_acumulada = (stock_anual + 1 ).product() - 1
ret_anual_acumulada
```

```python
ret_anual_promedio = (ret_anual_acumulada + 1) ** (1/stock_anual.shape[0]) - 1
ret_anual_promedio
```

```python
vol = stock_anual.std()
vol
```

```python
## Activo libre de riesgo
ALR = yf.download( '^IRX', start = "2017-10-01", end = "2024-10-02")
ALR_adj= ALR['Adj Close']
ALR_adj
```

```python
# Retornos trimestrales
ALR_anual = ALR_adj.resample('Q').apply(lambda x: x.iloc[0])/100/4
ALR_anual
```

```python
ALR_acumulada = (ALR_anual + 1 ).product() - 1
print(f'la rentabilidad acumuada es:{ALR_acumulada}')
```

```python
ARL_promedio_trimestral = (ALR_acumulada + 1) ** (1/ALR_anual.shape[0]) - 1
print(f'la rentabilidad  promedio trimestral es:{ARL_promedio_trimestral}')
```

```python
ARL_rent_mes = (ARL_promedio_trimestral + 1) ** (1/3) - 1
print(f'la rentabilidad mensual es:{ARL_rent_mes}')
```

```python
ARL_rent_año= (ARL_promedio_trimestral+1)**(4)-1
print(f'la rentabilidad anual es:{ARL_rent_año}')
```

```python
excess  = ret_anual_promedio - ARL_rent_año
excess 
```

EXCESS = Cuanto a superado EL activo, frente a la ALR

```python
ratio = excess/vol
print(f'El Ratio de Sharpe anual es: {ratio}')
```

Por cada unidad de volatilidad, vas a ganar BBVA 0.171348 comparado con el ALR

```python
import kit
```

```python
energia = ['REP.MC', 'IBE.MC','ELE.MC','NTGY.MC']
financiero = ['SAB.MC', 'SAN.MC','BKT.MC','BBVA.MC','CABK.MC','UNI.MC','MAP.MC']
stock_energia_adj = yf.download(energia, start="2022-02-18", end="2022-03-02")["Adj Close"]
stock_energia_adj
```

```python
retornos_ene = stock_energia_adj.pct_change().dropna()
retornos_ene
```

```python
f =retornos_ene.mean(axis=1)
f
```

```python
plt.style.use("default")

fig = plt.figure(figsize = (10,5))
ax = f.plot(kind="bar")
ax.set_title("Rendimientos medios")
ax.set_xlabel("Fecha")
ax.axhline(0, color="black", linewidth=1.0, alpha=0.5)
plt.show()
```

```python
financiero_df = yf.download(financiero, start ="2022-02-18", end ="2022-03-02")['Adj Close']
financiero_df
```

```python
ren_finan = financiero_df.pct_change().dropna()
ren_finan
```

```python
# Financiero
mean_finan = ren_finan.mean(axis=1)
mean_finan
```

```python
general =pd.concat([f,mean_finan],axis=1, keys=["Energia","Financieros"])
general
```

```python
plt.style.use("default")

fig = plt.figure(figsize = (10,5))
ax = general.plot(kind="bar")
ax.set_title("Rendimientos medios")
ax.set_xlabel("Fecha")
ax.axhline(0, color="black", linewidth=1.0, alpha=0.5)
plt.show()
```
