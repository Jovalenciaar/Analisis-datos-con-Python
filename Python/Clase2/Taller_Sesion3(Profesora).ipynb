```python
import pandas as pd
import numpy as np
import yfinance as yf
import kit  
import datetime
```

```python
stock_2 = ["ANA","ACX","ACS","AENA","AMS","BBVA","SAB",
"SAN","BKT","CABK","CLNX","ENG","ELE","FER","GRF","IBE","IDR","ITX","IAG","MAP","TL5",
"MEL","MRL","NTGY","RED","REP","SGRE","TRE","TEF","VIS"]
```

```python
df2 = yf.download(stock_2)
df2_ad = df2[['Adj Close']]
df2_ad.columns = df2_ad.columns.get_level_values(1)
df2_ad.index = df2_ad.index.tz_localize(None)
df2_ad.index = df2_ad.index.strftime('%Y-%m-%d')
ret_df2_ad = df2_ad.pct_change()
ret_df2_ad.fillna(0.0001,inplace=True)
ret_df2_ad.head()
```

```python
# Estamos calculando la desviación estándar para cada una de las columnas, sustituyendo los NaN por 0.001

for i,  sc in enumerate (ret_df2_ad.columns):
    STD = ret_df2_ad[sc].std()
    print(f"La Desviación estándar de {sc} es {STD}")
```

## Dividendos


```python

dividendos_info = {}

for i in stock_2:

    stock = yf.Ticker(i)
    dividendos = stock.dividends
    dividendos_info[i] = dividendos

    print(i,dividendos_info)
```

## Splits


```python
split_info = {}

for i in stock_2:
    stock = yf.Ticker(i)
    split = stock.splits
    split_info[i] = split

    print(i,split_info)
```
