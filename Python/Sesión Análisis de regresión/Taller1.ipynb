```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import yfinance as yf
```

```python
df = pd.read_csv('C:/Users/Lenovo/Programacion/Python/BBDD/bolsa_2021.csv', sep=",", index_col=0)
df
```

```python
stock = df[["Cie", "Ibez", "Enagas", "Santander"]]
retornos_stock = stock.pct_change().dropna()
retornos_stock
```

```python
interes = df[["interes"]]["20/10/2020":"18/10/2021"]
interes.dropna()
y =interes/36000
y
```

```python
x= pd.concat([retornos_stock, y], axis=1)
x
```

```python

x["pCie"] = x["Cie"] - x["interes"]
x["pIbez"] =  x["Ibez"] -x["interes"] 
x["pEnegas"] =  x["Enagas"] - x["interes"]
x["pSant"] = x["Santander"] - x["interes"]
```

```python
x
```

```python
est = smf.ols(formula='pCie ~ pIbez', data=x) # smf.ols modelo de regresion lineal simple (ols)
res = est.fit()
print(res.summary())
est = smf.ols(formula='pEnegas ~ pIbez', data=x)
res=est.fit()
print(res.summary())
est = smf.ols(formula='pSant ~ pIbez', data=x)
res=est.fit()
print(res.summary())
```
