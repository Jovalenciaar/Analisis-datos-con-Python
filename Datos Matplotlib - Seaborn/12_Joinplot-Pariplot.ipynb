```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = sns.load_dataset('tips')
t.head(3)
```

## Joinplot() - Pairplot()

These two funtion helps us to create different types of charts simultaneously

- Joinplot shows group charts on the same figure, and also shows the best way to express the datas.

- I may change the type of plot with the command "kind".

- With the statement "marginal" I may modify the charts that are on the main chart.

- Pairplot works for relationship of numeric variables.

- With Pariplot we can create groups charts, but in different figures.

```python
sns.jointplot(data=t, x="total_bill", y="tip", hue="sex")
```

```python
sns.jointplot(data=t, x="total_bill", y="tip", hue="sex",marginal_ticks=True)
```

```python
sns.pairplot(data=t)
```

```python
sns.pairplot(data=t, hue="sex")
```
