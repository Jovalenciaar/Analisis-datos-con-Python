```python
import matplotlib.pyplot as plt
import seaborn as sns
```

We are gonna to import a dataset from Seaborn in this case "titanic"

```python
titanic = sns.load_dataset("titanic")
titanic.head(4)
```

```python
sns.countplot(data= titanic, y="survived", hue="sex")
plt.show()
```

```python
sns.stripplot(data= titanic, x="survived",y="age", hue="sex", dodge=True)
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(data= titanic, x="survived",y="age", hue="sex", dodge=True)
```

```python
sns.swarmplot(data= titanic, x="survived",y="age", hue="sex", dodge=True)
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data= titanic, x="survived",y="age", hue="sex", dodge=True)
sns.stripplot(data= titanic, x="survived",y="age", hue="sex", dodge=True, color="0")
```

```python
sns.violinplot(data= titanic, x="survived",y="age", hue="sex", dodge=True)
```

```python
sns.violinplot(data= titanic, x="survived",y="age", hue="sex", split=True)
```

```python
sns.catplot(data= titanic,y="age", hue="sex",col="pclass")
## Col= nos ayuda a dividir por otra categoria nuestro grafico
```
