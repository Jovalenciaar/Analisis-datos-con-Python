```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
t = sns.load_dataset('tips')
t
```

```python
sns.scatterplot(data=t, x="total_bill", y="tip", hue="day", style="time", size="size", alpha=0.5)
plt.legend(loc="center", bbox_to_anchor=(1.11,0.5)) ## bbox_to_anchor= position
```

```python
sns.lineplot(data=t, x="total_bill", y="tip", hue="day", style="time", size="size", alpha=0.5)
```

```python
sns.relplot(data=t, x="total_bill", y="tip", hue="day", style="time", size="size", alpha=0.5)
```
