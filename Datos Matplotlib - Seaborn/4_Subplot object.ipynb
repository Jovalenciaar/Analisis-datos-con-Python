```python
import matplotlib.pyplot as plt
import numpy as np  
```

The before section we can see the object oriented method, but as happens at other times, we need to have multiple charts in one figure., Here is when Subplot shows up to help.


```python
x = np.linspace(7, 10*np.pi, 132)
y = np.cos(x)
z = np.tan(y)
q = np.arctan(x)
plt.style.use("default")
```

```python
## Define our figure and axes, but here we need to speficy row's and colum's number.
fig, axes = plt.subplots(nrows=1,ncols=2)
## Now we have a figure with 1 row and 2 columns, but to make reference to what axes we need,
    ##is neccesaary to indicate by position.
axes[0].plot(y)
axes[1].plot(x)
```

```python
## Define our figure and axes, but here we need to speficy row's and colum's number.
    ##We also can define the axes of this way.
fig, (axes1, axes2) = plt.subplots(nrows=1,ncols=2)
## Now we have a figure with 1 row and 2 columns, but to make reference to what axes we need,
    ##is neccesaary to indicate by position.
axes1.plot(y)
axes2.plot(x)
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(y,'r')
axes[0,1].plot(x)
axes[1,0].plot(z,'olive')
axes[1,1].plot(q,'green')
fig.suptitle('General graphic')
```

```python
fig, ((axes12, axes13),(axes14,axes15)) = plt.subplots(2,2)

axes12.plot(y,'r')
axes12.set_title('X')
axes13.plot(x)
axes13.set_title('Sin(x)')
axes14.plot(z,'olive')
axes14.set_title("Tan(y)")
fig.subplots_adjust(hspace=0.5)
axes15.plot(q,'green')
fig.suptitle('General graphic')
fig.subplots_adjust(wspace=0.5)
fig.subplots_adjust(right= 1.3)

```
