```python

import seaborn as sns
import matplotlib.pyplot as plt
```

As in Pyplot, in seaborn we need to import Seaborn with the command "SNS". and the syntasis is the same like in pyplot. 

```python
sns.barplot(x=['A', 'B', 'C'], y=[1,3,6]) ## we have to define the label x and y
plt.show()
```

## SET()

Set() helps us to define the style of chart

```python
sns.set(style="white", font="verdana", font_scale=2 )
## style= Style for chart
## pallete = intensity of color
## Font = Fuente
## Font_scale = size of letter
sns.barplot(x=['A', 'B', 'C'], y=[1,3,6],color="darkseagreen",edgecolor = "black") ## we have to define the label x and y

plt.show()
```
