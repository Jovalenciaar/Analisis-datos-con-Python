```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

Whe we use the object oriented method, can personalize each object that we have in the figure. It's a little more difficult but we create a lot of things. EX:


```python
## We have to creat a Obj where we can save our figure or chart.
fig = plt.figure()
## we need to create the chart in these cases we name like Axes
## Axes = Object (Fig) and we add axes
    ## The two first conditions are the position of the chart in our figure(0.1,0.1)
    ## The other two cond are the size of the chart (0.5,0.9)
axes = fig.add_axes([0.1,0.1,0.5,0.9])
## To show our chart we only need to write axes.plot(Indicate the parameters)
axes.plot(x,y,"deepskyblue")
axes.set_title('Graphic X-Y')
```
