```python
import matplotlib.pyplot as plt
import numpy as np
```

With the command or statement "plt.style.use( ), we can define the style of the chart that we wanna have.
This is neccesary to write in the beginnig of the proccess.

```python
x = np.linspace(2,440, 30)
print(plt.style.available)
```

```python
plt.style.use("seaborn-v0_8-ticks")
fig, axes = plt.subplots(figsize=(8,8))
## Alpha = transparencia
## Linewidth = width
## Linestyle = style
## Marker = marker of line
axes.plot(x,x*2, color= "green", alpha = 0.4)
axes.plot(x,x*3, color = 'fuchsia', linewidth = 3)
axes.plot(x,x*4, linestyle = '-.', marker = '3', markerfacecolor ="r")
```

```python
fig, axes = plt.subplots(figsize=(8,8))
axes.plot(x,x*2,'r--')
axes.plot(x,x*3,'b->')
axes.plot(x,x*4, 'y2')
```
