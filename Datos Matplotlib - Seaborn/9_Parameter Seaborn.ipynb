```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Seaborn has liberry or dataset included, so we can use them with the statement "load_dataset("")

dataset = tips- iris - titanic - flights

Seaborn helps us to identife which chart is better for each scenary

```python
t = sns.load_dataset("tips")
t
```

The way to create a chart is as follows:

sns.tipo_de_grafico(data= nuestro Dataframe, x= Que queremos en el eeje x, y= Que queremos en el eje Y, hue= manera de como queremos distribuir la informacion)

Ex:

sns.lmplot(data= t, x="total_bill", y = "tip", hue='sex')

```python
sns.displot(data= t, x="total_bill") ## data = our data , x= xlabel y = ylabel
# we have to define our data or datframe, and indicate which values we wanna have in the label X and Y
plt.show()
```

```python
sns.kdeplot(data= t, x="total_bill") ## data = our data , x= xlabel y = ylabel
# we have to define our data or datframe, and indicate which values we wanna have in the label X and Y
plt.show()
```

```python
sns.lmplot(data= t, x="total_bill", y = "tip") ## data = our data , x= xlabel y = ylabel
# we have to define our data or datframe, and indicate which values we wanna have in the label X and Y
plt.show()
```
