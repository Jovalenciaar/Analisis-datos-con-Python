## Subplot()

As Pyplot, subplot helps us to create and show charts, but what happends when we wanna have more than one chart in the presentation, here is when Subplot show up. With Subplot we can create more than one chart in the presentation.

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
## Here we have to indicate the numbers of chart we want.
## First condit is  = row's number
## Secound = column's number
## Thirth = What chart we can show, inb this case this is the first chat
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y,x)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.plot(y,x,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,linewidth=5)
plt.show()
```
