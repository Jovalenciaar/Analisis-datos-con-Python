```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Scatter

```python
X = np.random.randint(1,30,200)
Y = np.random.randint(X) / 3
O = np.random.randint(1,300,200)
color = np.random.rand(200) ## Create random color for every point
plt.scatter(X,Y,s=O,marker="p", alpha=0.5, c =color) ## Aplha helps us to the transparency
plt.colorbar()
plt.show()
```
