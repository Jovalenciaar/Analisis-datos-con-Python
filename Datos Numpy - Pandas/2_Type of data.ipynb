FLOAT

We have different types of data such as int = integer or float = decimals, for knowing what type of data we have, we only need to type the following statement "dtype", this way we can know what type of data we are working with, but if we need to change the type of data, juswe only need to indicate it the array that we have, what typye of data we want.

```python
arr2 = np.array([2,3,4,5], dtype = float)
arr2.dtype
```

```python
arr2
```

If we want to change an array definitely that we already have previously, we only have to write "name of array.astype(np.float64)", this way we can change the type of data ex:

```python
new = np.array([2,4,3,6])
new.dtype    
```

```python
new = new.astype(np.float64)
new
```

BOOL

The statement "bool_" helps us to change the array to type of data bool = false or true Ex:

```python
new2 = np.array([1,0,9,3,4,0])
new2
```

```python
new2 = new2.astype(np.bool_)
new2
```

STRING

The data type "string" mean that each character in our array is a "Text" Ex:

```python
tex = np.array([1,2,4,'e', 5])
tex = tex.astype(np.string_)
tex
```
