```python
car = pd.read_csv('C:/Users/Lenovo/Programacion/Base de datos/CarPricesPrediction.csv')
car.drop('Unnamed: 0', axis=1, inplace=True)
car["Price"] = car['Price'].astype(int)
car
```

## Pivot_Table()

This command helps us to create a new table, such as dynamic table in Excel, with this command we can create our own table, expecifying de index, columns and values. Ex:


```python
car.pivot_table(index="Make", columns='Year', values="Price")
```

## Melt()

This command turn around the columns of our dataframe in rows

```python
car.melt(id_vars='Year', value_vars='Price')
```
